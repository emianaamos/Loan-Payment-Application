```python
df1=df1.drop(labels='Loan_ID', axis=1)
df1
```

```python
df1.Property_Area.unique()
```

```python
df1['Gender']=pd.get_dummies(df1['Gender'])
df1['Married']=pd.get_dummies(df1['Married'])
df1['Self_Employed']=pd.get_dummies(df1['Self_Employed'])
df1['Loan_Status']=pd.get_dummies(df1['Loan_Status'])
df1['Education']=pd.get_dummies(df1['Education'])

from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df1['Property_Area']=label_encoder.fit_transform(df1['Property_Area'])

df1
```

```python
data=df1.drop('Loan_Status', axis=1)
target=df1.Loan_Status
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
%%time
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```
